```python
X = df.drop('target',axis=1)
y = df['target']
```

```python
#Spliting dataset into Training and Testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, train_size=0.8, random_state=42)

```

```python
#As we know our target is either Yes or No, so we do classification

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
#This is for Evaluating the model
from sklearn.metrics import accuracy_score, classification_report, recall_score
```

```python
log_reg = LogisticRegression()
knc = KNeighborsClassifier()
svc = SVC()
dtc = DecisionTreeClassifier()
```

```python
log_reg.fit(X_train, y_train)
knc.fit(X_train, y_train)
svc.fit(X_train, y_train)
dtc.fit(X_train, y_train)
```

## Here we start evaluating the model

```python
y1_pred = log_reg.predict(X_test)
y2_pred = knc.predict(X_test)
y3_pred = svc.predict(X_test)
y4_pred = dtc.predict(X_test)
```

## Now we find out the Accuracy, and recall_score of our models

```python
acc_log_reg = accuracy_score(y_test, y1_pred)
print("Accuracy of Logistic Regression is: ", acc_log_reg)
```

```python
re_log_reg = recall_score(y_test, y1_pred)
print("Recall of Logistic Regression is: ", re_log_reg)
```

```python
acc_knc = accuracy_score(y_test, y2_pred)
print("Accuracy of KNeighbors Classifier is: ", acc_knc)
```

```python
re_knc = recall_score(y_test, y2_pred)
print("Recall of KNeighbors Classifier is: ", re_knc)
```

```python
acc_svc = accuracy_score(y_test, y3_pred)
print("Accuracy of SVC is: ", acc_svc)
```

```python
re_svc = recall_score(y_test, y3_pred)
print("Recall of SVC is: ", re_svc)
```

```python
acc_dtc = accuracy_score(y_test, y4_pred)
print("Accuracy of Decision Tree Classifier is: ", acc_dtc)
```

```python
re_dtc = recall_score(y_test, y4_pred)
print("Recall of Decision Tree Classifier is: ", re_dtc)
```

```python
cm_log_reg = classification_report(y_test, y1_pred)
print("Classification Report of Logistic Regression is: \n", cm_log_reg)
```

```python
cm_knc = classification_report(y_test, y2_pred)
print("Classification Report of KNeighbors Classifier is: \n", cm_knc)
```

```python
cm_svc = classification_report(y_test, y3_pred)
print("Classification Report of SVC is: \n", cm_svc)
```

```python
cm_dtc = classification_report(y_test, y4_pred)
print("Classification Report of Decision Tree Classifier is: \n", cm_dtc)
```

```python
#Note: Only this cell code is AI generated

# Compare models based on Accuracy and Recall
results = pd.DataFrame({
    'Model': ['Logistic Regression', 'KNeighbors Classifier', 'SVC', 'Decision Tree Classifier'],
    'Accuracy': [acc_log_reg, acc_knc, acc_svc, acc_dtc],
    'Recall': [re_log_reg, re_knc, re_svc, re_dtc]
})

# Sort by Accuracy
results_accuracy = results.sort_values(by='Accuracy', ascending=False)
print("\nComparison by Accuracy:")
display(results_accuracy)

# Sort by Recall
results_recall = results.sort_values(by='Recall', ascending=False)
print("\nComparison by Recall:")
display(results_recall)

# Determine the best model based on Accuracy (you can choose other metrics as well)
best_model_accuracy = results_accuracy.iloc[0]
print(f"\nBest model based on Accuracy: {best_model_accuracy['Model']} with Accuracy: {best_model_accuracy['Accuracy']:.4f}")

# Determine the best model based on Recall
best_model_recall = results_recall.iloc[0]
print(f"\nBest model based on Recall: {best_model_recall['Model']} with Recall: {best_model_recall['Recall']:.4f}")
```
